```python
#Profit Mix
import pandas as pd
import matplotlib.pyplot as plt

#Converting Shipping Date collumn to datetime format
data['Shipping Date'] = pd.to_datetime(data['shipping date (DateOrders)'])

#Extract the year from Shipping Date
data['Year'] = data['Shipping Date'].dt.year

# Group data by Profit/Year/Department Name
profit_by_year_department = data.groupby(['Year', 'Department Name'])['Order Profit Per Order'].sum()

#Profit Mix pie chart
for year in data['Year'].unique():
    profit_for_year = profit_by_year_department[year]
    plt.figure(figsize=(8, 8))
    plt.pie(profit_for_year, labels=profit_for_year.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Profit Mix for Year {year}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```

```python
#Seasonality/Sales by Department each Year
import pandas as pd
import matplotlib.pyplot as plt

# Convert the 'shipping date (DateOrders)' column to datetime format
data['shipping date (DateOrders)'] = pd.to_datetime(data['shipping date (DateOrders)'])

# Extract the year and month from the 'shipping date (DateOrders)' column
data['Year'] = data['shipping date (DateOrders)'].dt.year
data['Month'] = data['shipping date (DateOrders)'].dt.month

# Group the data by year, month, and department and sum up the sales
sales_by_year_month_department = data.groupby(['Year', 'Month', 'Department Name'])['Sales'].sum().unstack()

# Plot separate line charts for each year
for year in data['Year'].unique():
    year_data = sales_by_year_month_department.loc[year]
    
    # Check if there is any data for the current year
    if not year_data.isnull().all().all():
        plt.figure(figsize=(12, 8))
        departments = year_data.columns
        
        for department in departments:
            plt.plot(year_data.index, year_data[department], marker='o', label=department)
        
        plt.title(f'Sales by Department for Year {year}')
        plt.xlabel('Month')
        plt.ylabel('Sales')
        plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
        plt.legend()
        plt.grid(True)
        plt.show()
```

```python
#Products we are losing money
import pandas as pd

# Group the data by 'Product Name' and calculate the average profit for each product
average_profit_by_product = data.groupby('Product Name')['Order Profit Per Order'].mean()

# Filter out the products with negative average profit
products_with_negative_profit = average_profit_by_product[average_profit_by_product < 0]

# Print the products with negative average profit
print("Products with negative average profit:")
print(products_with_negative_profit)
```

```python
#Low, Average, High Profit Ratio products
import pandas as pd

# Group the data by 'Product Name' and calculate the average Order Item Profit Ratio for each product
average_profit_ratio_by_product = data.groupby('Product Name')['Order Item Profit Ratio'].mean()
#Categorize each product based on their average profit margin
def categorize_profit_margin(avg_profit_ratio):
    if avg_profit_ratio < 0.05:
        return 'Low Profit Margin'
    elif 0.1 <= avg_profit_ratio <= 0.15:
        return 'Average Profit Margin'
    else:
        return 'High Profit Margin'

# Apply the categorization function to each product
product_categories = average_profit_ratio_by_product.apply(categorize_profit_margin)

# Print the categorized products
print("Categorized Products:")
print(product_categories)
```
